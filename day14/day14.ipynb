```python
BIT_LEN = 36
def pt1(lines):
    def apply_bitmask(value,bit_mask, bit_pos):
        #print(f'Value in: {value}')
        for idx,mask in enumerate(bit_mask):
            if(mask == 0): #set the bit to 1
                value = value & (~(2**(bit_pos[idx]-1)))
            elif(mask == 1): #set the bit to 1
                value = value | (2**(bit_pos[idx]-1))
            else:
                print(f"What? {mask}")
        #print(f'Value out: {value}')
        return value
    def get_bitmask(bit_mask_str):
        #get bit mask list and bit pos list
        bit_mask = []
        bit_pos = []
        for idx,char in enumerate(bit_mask_str):
            if(char != 'X'):
                bit_mask.append(int(char))
                bit_pos.append(BIT_LEN -idx)
        return bit_mask, bit_pos
    mem = {}
    bit_mask_str = ['X' for i in range(BIT_LEN)]
    bit_mask, bit_pos = get_bitmask(bit_mask_str)
    for line in lines:
        if 'mask' in line: #bit mask
            bit_mask_str = line.split(' ')[2]
            bit_mask, bit_pos = get_bitmask(bit_mask_str)
            #print(bit_mask,bit_pos)
        else:
            mem_loc,_,value = line.split(" ")
            mem_loc = int(mem_loc.replace('mem','').replace('[','').replace(']',''))
            value = int(value)
            mem[mem_loc] = apply_bitmask(value,bit_mask, bit_pos)
            #print(f'Mem: {mem_loc}, Value: {value}')
    #sum up memory values
    acc = 0
    for key in mem.keys():
        acc += mem[key]
    return acc
```
