```python
import math
with open('input.txt','r') as f:
    time = int(f.readline().strip())
    bus_slots = f.readline().strip().split(',')
    bus_ids = list(filter(lambda a: a != 'x', bus_slots))
    bus_slot_idxs = [bus_slots.index(x) for x in bus_ids]
    bus_ids = [int(x) for x in bus_ids]
```

```python
time,bus_ids,bus_slot_idxs
```

```python
#find next time
next_times = [math.ceil(time/id)*id for id in bus_ids]
delta_times = [next_time - time for next_time in next_times]
delta_times
```

```python
from operator import itemgetter
idx_min = min(enumerate(delta_times), key=itemgetter(1))[0]
bus_ids[idx_min],delta_times[idx_min]
```

```python
#Need t such that t mod 7 == 0 and (t+1) mod 13 = 0 and t+4 mod 59 == 0 and t+6 mod 31 == 0 and t+7 mod 19 == 0
```

```python
# t = 7*N, (t+1 = 13 * M), (t + 4 = 59 * L), (t+6 = 31 * Q), (t + 7 = 19 * R)...
#7*N +1 = 13 * M, 13*M-7*N=1
```

```python
def get_pt2(bus_ids,bus_idxs):
    t = 0
    mult = 1
    for i in range(len(bus_ids)-1):
        mult *= bus_ids[i]
        slot = bus_ids[i+1]
        idx = bus_idxs[i+1]
        while((t+idx) % slot != 0):
            t += mult
            print(mult,t)
    return t
```
