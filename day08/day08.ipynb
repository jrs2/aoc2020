```python
with open('input.txt','r') as f:
    lines = [line.strip() for line in f.readlines()]
```

```python
def parse_line(line):
    arg,num = line.split(' ')[0], int(line.split(' ')[1])
    #print(arg,num)
    if arg == 'nop':
        acc_mod = 0
        offset_mod = 1
    elif arg == 'jmp':
        acc_mod = 0
        offset_mod = num
    elif arg == 'acc':
        acc_mod = num
        offset_mod = 1
    else:
        print(f'Error with unexpected {arg}')
    return acc_mod,offset_mod
```

```python
def traverse_program(lines):
    acc = 0
    offset = 0
    lines_encountered = set()
    while True:
        if offset in lines_encountered:
            print(f'Same offset found: {offset} and acc: {acc}')
            break
        else:
            lines_encountered.add(offset)
        acc_mod, offset_mod = parse_line(lines[offset])
        acc += acc_mod
        offset += offset_mod
        if(offset == len(lines)):
            print(f'Found end with acc at : {acc}')
            break
    return offset,acc
traverse_program(lines)
```

```python

mod_offset = 0
#loop through and mod one line at a time and then mod back if we don't reach EOP
while(mod_offset < len(lines)):
    line = lines[mod_offset]
    arg,num = line.split(' ')[0], int(line.split(' ')[1])
    if(arg == 'nop'):
        lines[mod_offset] = lines[mod_offset].replace('nop','jmp')
    elif (arg == 'jmp'):
        lines[mod_offset] = lines[mod_offset].replace('jmp','nop')
    else:
        mod_offset += 1
        continue

    offset, acc = traverse_program(lines)
    
    if(offset == len(lines)):
        break
    if(arg == 'nop'):
        lines[mod_offset] = lines[mod_offset].replace('jmp','nop')
    elif (arg == 'jmp'):
        lines[mod_offset] = lines[mod_offset].replace('nop','jmp')
    mod_offset += 1
```
