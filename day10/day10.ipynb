```python
import pandas as pd
df = pd.read_csv('input.txt',header=None)
```

```python
df = df.append([df.max()[0]+3,0])
```

```python
df.sort_values(by=0,inplace=True)
```

```python
df['deltas'] = df - df.shift(1)
```

```python
df['deltas'].value_counts()[1.0] * df['deltas'].value_counts()[3.0]
```

```python
df[(df['deltas'] == 1.0) & (df['deltas'] == df['deltas'].shift(-1))]
```

```python
#can't remove any of the 3 deltas
#can't remove any of the 1 deltas with an subsequent 3
#let's count for adjacent double adjacent 1 deltas this is a 2 multiplier as there are 2 combinations (Null and (1))
count_2 = df[(df['deltas'] == 1.0) &
           (df['deltas'] == df['deltas'].shift(-1)) &
           (df['deltas'] != df['deltas'].shift(-2)) &
           (df['deltas'].shift(1) != 1.0)].count()[0]
count_2
```

```python
#let's count for adjacent triple adjacent 1 deltas this is a 4 multiplier as there are 4 (Null, (1), (2), (1,2))
#Valid options are Null, (1), (2), and (1,2)
count_3 = df[(df['deltas'] == 1.0) &
           (df['deltas'] == df['deltas'].shift(-1)) &
           (df['deltas'] == df['deltas'].shift(-2)) &
           (df['deltas'] != df['deltas'].shift(-3)) &
           (df['deltas'].shift(1) != 1.0)].count()[0]
count_3
```

```python
#let's count for adjacent quadruple adjacent 1 deltas, this is where it gets interesting
#you can only take out 2 of these at a time, so Null, (1), (2), (3), (1,2), (2,3), and (1,3) are valid options
#So 7 total
# 1
# 1
# 1
# 1
# 3
count_4 = df[(df['deltas'] == 1.0) & \
           (df['deltas'] == df['deltas'].shift(-1)) &
           (df['deltas'] == df['deltas'].shift(-2)) &
           (df['deltas'] == df['deltas'].shift(-3)) &
           (df['deltas'] != df['deltas'].shift(-4)) &
           (df['deltas'].shift(1) != 1.0)].count()[0]
count_4
```

```python
#check for 5 or greater
df[(df['deltas'] == 1.0) & \
           (df['deltas'] == df['deltas'].shift(-1)) &
           (df['deltas'] == df['deltas'].shift(-2)) &
           (df['deltas'] == df['deltas'].shift(-3)) &
           (df['deltas'] == df['deltas'].shift(-4)) ].count()[0]
```
