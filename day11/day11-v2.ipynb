```python
import pandas as pd
df=pd.read_csv('input.txt',header=None)
df = df[0].apply(lambda x: pd.Series(list(x)))
df
```

```python
EMPTY = 'L'
FULL = '#'
def check_surrounding_seats(input_df,row,col):
    temp = input_df.iloc[row,col]
    input_df.iloc[row,col] = '' #ignore actual seat cell
    #max (0,X) for left and upper edges (can handle ending edges)
    adj_block = input_df.iloc[max(0,(row-1)):(row+2),max(0,(col-1)):(col+2)]
    full_count = adj_block[adj_block == FULL].count().sum()
    input_df.iloc[row,col] = temp
    return full_count

def fill_seats(input_df):
    temp_copy = input_df.copy(deep=True)
    input_df
    for row_idx, row in input_df.iterrows():
        for col_idx, col in row.iteritems():
            if(temp_copy.iloc[row_idx,col_idx] == EMPTY):
                if(check_surrounding_seats(input_df,row_idx,col_idx) == 0):
                    temp_copy.iloc[row_idx,col_idx] = FULL
    return temp_copy

def remove_seats(input_df):
    temp_copy = input_df.copy(deep=True)
    for row_idx, row in input_df.iterrows():
        for col_idx, col in row.iteritems():
            if(temp_copy.iloc[row_idx,col_idx] == FULL):
                if(check_surrounding_seats(input_df,row_idx,col_idx) >= 4):
                    temp_copy.iloc[row_idx,col_idx] = EMPTY
    return temp_copy
```

```python
count = 0
while(True):
    print(f'Iteration: {count}')
    s1 = fill_seats(df)
    s2 = remove_seats(s1)
    mismatches = (s2 != s1).sum().sum()
    if(mismatches == 0):
        break
    df = s2
    count += 1
    print(f'Mismatches: {mismatches}')
```
