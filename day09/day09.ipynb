```python
import pandas as pd
df = pd.read_csv('input.txt',header=None)
```

```python
#just loop through and do pairs
preamble_len = 25
for i in range(1,len(df)):
    df[i] = None
    df.iloc[(i):(i+preamble_len),i] = df.iloc[(i-1):(i+preamble_len),0] + df.iloc[i-1,0]
```

```python
df.iloc[0:25,0:25]
```

```python
#now loop through the 25x25 blocks to see if the value is in there
for i in range(25,len(df)):
    if df.iloc[i,0] not in df.iloc[(i-preamble_len):i,(i-preamble_len):i].values:
        print(f'No match found {df.iloc[i,0]} index: {i}')
        value = df.iloc[i,0]
        index = i
        break
```

```python
#now loop through and do running sum (maybe should have used pandas.running(sum))
for i in range(1,index):
    df[i] = 0
    if(i==1):
        df.iloc[i,1] = df.iloc[(i-1),0] + df.iloc[i,0]
    else:
        df.iloc[i,1:(i+1)] = df.iloc[(i-1),0:i].values + df.iloc[i,0]
    if(value in df.iloc[i,0:i].values):
        print(i)
        new_index = i
        break
```

```python
for i in range((new_index-30),(new_index+1)):
    if(value - df.iloc[i:(new_index+1),0].sum()) == 0:
        print(df.iloc[i:(new_index+1),0].min()+df.iloc[i:(new_index+1),0].max())
        break
```
