```python
import re
with open('input.txt','r') as f:
    lines = [line.strip() for line in f.readlines()]
```

```python
bag_map = {}
for line in lines:
    outter_bag,inner_bags = line.split('contain')
    outter_bag,inner_bags = outter_bag.replace('bags','').replace('bag','').strip(),inner_bags.replace('.','').strip().split(',')
    for inner_bag in inner_bags:
        base_inner_bag = re.sub("\d+", "", inner_bag).replace('bags','').replace('bag','').strip()
        if(base_inner_bag in bag_map.keys()):
            bag_map[base_inner_bag].append(outter_bag)
        else:
            bag_map[base_inner_bag] = [outter_bag]
        #print(base_inner_bag)
    #print(outter_bag)
```

```python
bag_map['shiny gold']
```

```python
def find_all_outter_bags(bag_type):
    all_bags = set()
    if(bag_type in bag_map.keys()):
        all_bags = set(bag_map[bag_type])
        for bag in all_bags:
            #print(bag)
            new_bags = find_all_outter_bags(bag)
            #print(f'New bags: {new_bags}')
            all_bags = all_bags.union(new_bags)
            #print(f'All bags: {all_bags}')
    return all_bags
```

```python
bag_map = {}
bag_counts = {}
for line in lines:
    outter_bag,inner_bags = line.split('contain')
    outter_bag,inner_bags = outter_bag.replace('bags','').replace('bag','').strip(),inner_bags.replace('.','').strip().split(',')
    for inner_bag in inner_bags:
        base_inner_bag = re.sub("\d+", "", inner_bag).replace('bags','').replace('bag','').strip()
        #print(inner_bag)
        if('no other bag' in inner_bag):
            number_inner_bag = 0
        else:
            number_inner_bag = int(re.findall("(\d+)",inner_bag)[0])
        #print(number_inner_bag)
        if(outter_bag in bag_map.keys()):
            bag_map[outter_bag].append(base_inner_bag)
            bag_counts[outter_bag].append(number_inner_bag)
        else:
            bag_map[outter_bag] = [base_inner_bag]
            bag_counts[outter_bag] = [number_inner_bag]
        #print(base_inner_bag)
bag_counts
```

```python
def find_all_inner_bags(bag_type):
    bag_count = 0
    if(bag_type in bag_map.keys()):
        next_bags = bag_map[bag_type]
        next_counts = bag_counts[bag_type]
        #print(next_bags,next_counts)
        for idx,bag in enumerate(next_bags):
            bag_count += next_counts[idx]
            if(next_counts[idx] != '0'):
                bag_count = bag_count + next_counts[idx] * int(find_all_inner_bags(bag))
        #print(bag_count)
    return bag_count
find_all_inner_bags('shiny gold')
```
